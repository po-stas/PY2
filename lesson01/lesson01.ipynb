1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

В моём понимании максимизация Precission необходима когда цена false-positive ошибки высока, 

- Скажем в задачах фильтрации спама в электронной почте, где перспектива не распознать какое-то количество спам писем не так страшна, как отправка в спам важной корреспонденции.
- Возможно Precission так-же важнее в рекомендательных системах - например, по подбору потенциально интересных фильмов (хотя, полагаю, такие задачи вообще может быть не про классификацию - а скорее про кластеризацию)

По поводу максимизации Recall - это в случаях когда мы можем смириться с какой-то долей ошибок, в угоду широте охвата.. 

- Мне видится, что какие-нибудь системы предсказания наличия у человека болезни, типа "ковид19", под подозрение попадает даже те, кто возможно и не болен, но это лучше, чем упустить реально больного.
- Похожая история наверное может быть в банковских предсказаниях кредитных рейтингов, в моем понимании, банку наверное лучше перестраховаться, чем пропустить неплательщика. (хотя это общие рассуждения. В этой области я, честно, не силён)

2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Единая метрика в техническом плане удобнее двух величин. Что касается среднего от точности и полноты - F метрика позволяет через коэффициент $\beta$ гибко настраивать, что нам важнее, precission или recall и насколько важнее.
