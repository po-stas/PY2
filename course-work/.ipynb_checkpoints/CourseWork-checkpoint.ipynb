```python
def standardize(df, feature_names, target):
    test_df = df.copy()
    test_df.reset_index(inplace=True)
    # scaler = StandardScaler()
    # features_to_stand = test_df[feature_names].select_dtypes(include='float64').columns.tolist()
    # stand_features = scaler.fit_transform(test_df[features_to_stand])
    
    # test_df[features_to_stand] = pd.DataFrame(stand_features, columns=features_to_stand)
    X = test_df.drop([target], axis=1)
    y = pd.Series(test_df[target].values)
    return X, y
```

```python
def real_standardize(df, feature_names, target):
    test_df = df.copy()
    test_df.reset_index(inplace=True)
    scaler = StandardScaler()
    features_to_stand = test_df[feature_names].select_dtypes(include='float64').columns.tolist()
    stand_features = scaler.fit_transform(test_df[features_to_stand])
    
    test_df[features_to_stand] = pd.DataFrame(stand_features, columns=features_to_stand)
    X = test_df.drop([target], axis=1)
    y = pd.Series(test_df[target].values)
    return X, y
```

```python
def prepare_current_loan(df):
    new_features = ['Bankruptcies-Number of Open Accounts', 'Current Credit Balance-Purpose_debt consolidation', 'Bankruptcies-Months since last delinquent', 'Purpose_debt consolidation-Purpose_wedding', 'Term_Short Term-Purpose_medical bills', 'Number of Open Accounts-Number of Credit Problems', 'Bankruptcies-Purpose_take a trip', 'Tax Liens-Home Ownership_Rent', 'Home Ownership_Rent-Monthly Debt', 'Purpose_renewable energy-Maximum Open Credit', 'Months since last delinquent-Purpose_business loan', 'Purpose_buy a car-Purpose_business loan', 'Months since last delinquent-Tax Liens']
    df_with_new_features = df.copy()
    columns = []
    for new_feature in new_features:
        a, b = new_feature.split('-')
        df_with_new_features[new_feature] = df_with_new_features[a] * df_with_new_features[b]
        columns.append(new_feature)

    feature_names = ['Annual Income', 'Years in current job', 'Tax Liens', 'Number of Open Accounts', 'Years of Credit History', 'Maximum Open Credit', 'Number of Credit Problems', 'Bankruptcies', 'Current Credit Balance', 'Monthly Debt', 'Credit Score', 'Months since last delinquent']
    feature_names.extend(columns)
    X, y = standardize(df_with_new_features, feature_names, 'Current Loan Amount')
    return X, y
```
