```python
import re
def to_digit(source):
    return int(re.findall(r'\d+', source)[0])

data_one_hot['Years in current job'] = data_one_hot.apply(lambda x: to_digit(x['Years in current job']), axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
def standardize(df, feature_names, target):
    test_df = df.copy()
    test_df.reset_index(inplace=True)
    scaler = StandardScaler()
    features_to_stand = test_df[feature_names].select_dtypes(include='float64').columns.tolist()
    stand_features = scaler.fit_transform(test_df[features_to_stand])
    
    test_df[features_to_stand] = pd.DataFrame(stand_features, columns=features_to_stand)
    X = test_df.drop([target], axis=1)
    y = pd.Series(test_df[target].values)
    return X, y
```

```python
# Попробуем предсказать Current Loan Amount
feature_names = ['Annual Income', 'Years in current job', 'Tax Liens', 'Number of Open Accounts', 'Years of Credit History', 'Maximum Open Credit', 'Number of Credit Problems', 'Bankruptcies', 'Current Credit Balance', 'Monthly Debt', 'Credit Score', 'Months since last delinquent']
X, y = standardize(final_temp_train, feature_names, 'Current Loan Amount')
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None)
# pd.options.display.float_format = '{:.0f}'.format
X.head()
```

```python
gb_model = GradientBoostingRegressor()
parms = {'n_estimators': [50, 100, 200], 'max_depth': [2, 3, 4, 5]}
gs = GridSearchCV(gb_model, parms, scoring='r2', cv=KFold(n_splits=5, random_state=100, shuffle=True), n_jobs=-1)
gs.fit(X, y)
```

```python
# Попробуем синтетические признаки
import random
random.seed(101)
max_score = 0
for features in range(3, 50):
    f_set = set()
    columns = []
    test_final_temp_train = final_temp_train.copy()
    for feature in range(1, features):

        f1 = random.choice([x for x in final_temp_train.columns if x != 'Current Loan Amount'])
        f2 = random.choice([x for x in final_temp_train.columns if x != 'Current Loan Amount'])
        if f1 == f2:
            continue

        f_set.add((f1, f2))

    for a, b in f_set:
        column = a + '-' + b
        test_final_temp_train[column] = test_final_temp_train[a] * test_final_temp_train[b]
        columns.append(column)

    feature_names = ['Annual Income', 'Years in current job', 'Tax Liens', 'Number of Open Accounts', 'Years of Credit History', 'Maximum Open Credit', 'Number of Credit Problems', 'Bankruptcies', 'Current Credit Balance', 'Monthly Debt', 'Credit Score', 'Months since last delinquent']
    feature_names.extend(columns)
    X, y = standardize(test_final_temp_train, feature_names, 'Current Loan Amount')
    gb_model = GradientBoostingRegressor()
    parms = {'n_estimators': [100, 150, 200], 'max_depth': [2, 3, 4, 5]}
    gs = GridSearchCV(gb_model, parms, scoring='r2', cv=KFold(n_splits=5, random_state=100, shuffle=True), n_jobs=-1)
    gs.fit(X, y)
    if gs.best_score_ > max_score:
        max_score = gs.best_score_
        print('{} added features'.format(len(columns)))
        print(columns)
        print(gs.best_score_)
        print(gs.best_params_)
        print('==============================')
```
